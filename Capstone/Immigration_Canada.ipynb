## Data Cleaning Practice - Immigration to Canada from 2015 - 2021

### Introduction 
In this Notebook I have taken data from the Government of Canada Immigration files at : https://open.canada.ca/data/en/dataset/f7e5498e-0ad8-4417-85c9-9b8aff9b9eda
<br>The purpose of this notebook is to use the dataset to practice data cleaning using pandas and plotting simple graphs using matplotlib 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_excel("Immigrations_by_CountryofOrigin.xls")
```

```python
data1.head(10)
```

```python
data1.columns = data1.iloc[3]
```

```python
data1.drop(labels=3, axis=0, inplace=True)
```

```python
data1.drop(labels="new", axis=1, inplace=True)
```

```python
data1.columns.values[1] = "Jan"
```

```python
data1.head(8)
```

```python
data1.head(5)
```

```python
years = {}
first=1;
second=18;
for i in range(15, 21):
    print(i)
    years["years_" + str(2000 + i)] = data1.iloc[:, np.r_[0, first:second]]
    first = second;
    second = second + 17;
```

```python
df18 = years["years_2018"]
df18.drop(labels=[0,1,2], axis=0, inplace=True)
```

```python
df17 = years["years_2017"]
df17.drop(labels=[0,1,2], axis=0, inplace=True)
```

```python
df17.drop(labels=[0,1,2], axis=0, inplace=True)
```

```python
df17
```

```python
df17.dropna(axis=0)
```

```python
df17.rename(columns={"temp":"Year Total"}, inplace=True)
```

```python
df17.head(2)
```

```python
df17.replace("--", 0, inplace = True)
```

```python
df17.dropna(inplace=True)
```

```python
df17
```

```python
df17.reset_index(drop=True, inplace=True)
```

```python
df17.head(10)
```

```python
df17.set_index('Location', inplace=True)
```

```python
df17.head(8)
```

### NewfoundLand and Labrador Data

```python
nfld_17 = df17.loc["St. John's":"Other - Newfoundland and Labrador", :]
```

```python
nfld_17.drop(labels="Total", axis = 0, inplace=True)
```

```python
## Bar Graph
nfld_17.plot.barh(y='Year Total', color="Maroon")
plt.title("Immigration to Newfoundland and Labrador in 2017")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(np.arange(len(nfld_17.index)), nfld_17["Year Total"], align='center', color="maroon")
ax.set_yticks(np.arange(len(nfld_17.index)))
ax.set_yticklabels(nfld_17.index)
ax.invert_yaxis()
ax.set_xlabel('Number of Immigrants')
ax.set_title("Immigration to Newfoundland and Labrador in 2017")
plt.show()
```
